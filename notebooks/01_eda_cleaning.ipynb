```python
print("Retailers:", df["customerId"].nunique())
print("Products:", df["skuNumber"].nunique())
print("Orders:", df["orderNumber"].nunique())
print("Categories:", df["category"].nunique())
```

```python
df_clean = df[
    df["orderStatus"].isin(
        ["Delivered", "PartiallyDelivered"]
    )
].copy()

print(df_clean.shape)
```

```python
df_clean["effective_qty"] = df_clean.apply(
    lambda row:
    row["deliveredQty"]
    if row["orderStatus"] == "PartiallyDelivered"
    else row["orderQty"],
    axis=1
)

df_clean[
    ["orderQty",
     "deliveredQty",
     "effective_qty"]
].head()
```

```python
df_clean = df_clean[
    df_clean["effective_qty"] > 0
]

print(df_clean.shape)
```

```python
df_clean["createdAt"] = pd.to_datetime(
    df_clean["createdAt"]
)

print(
    df_clean["createdAt"].min()
)

print(
    df_clean["createdAt"].max()
)
```

```python
df_clean.to_parquet(
    "../data/processed/orders_clean.parquet",
    index=False
)

print("Saved successfully")
```
